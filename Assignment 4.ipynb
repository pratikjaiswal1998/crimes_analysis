```python
crime = crime.drop(['ID'],axis =1)
crime = crime.drop(['Case Number'],axis =1)
```

```python
crime.shape
```

```python
crime['date1'] = pd.to_datetime(crime['Date'])
```

```python
crime.shape
```

```python
crime['Year'] = crime['date1'].dt.year
crime['Month']= crime['date1'].dt.month
crime['Day'] = crime['date1'].dt.day
crime['Hour'] = crime['date1'].dt.hour
crime.head()
```

```python
print(crime['Primary Type'].unique())
```

```python
((crime['Primary Type'].value_counts()/len(crime)) *100).plot.bar(width = 0.5)
```

```python
hour_wise = crime.groupby(['Hour']).count()
hour_wise
```

```python
hour_wise[['Primary Type']].plot.bar()
```
